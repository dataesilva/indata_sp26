```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
spot = pd.read_csv("spotify-2023.csv", encoding = 'latin1')
spot.head()
```

```python
spot.columns, spot.keys()
```

```python
hist1 = sns.histplot(data = spot, x = 'bpm', kde = True)
# hist1.plot()
plt.show()
```

```python
releasemonth_group = spot.groupby('released_month').count()['track_name']
releasemonth_group
```

```python
releasemonth_values = releasemonth_group.values
releasemonth_index = releasemonth_group.index
```

```python
releasemonth_index = releasemonth_group.index.astype('category')
```

```python
cat1 = sns.catplot(x = releasemonth_index, y = releasemonth_values, kind = 'bar',
                   hue = releasemonth_index, palette = "Spectral", legend = False)
plt.show()
```

```python
releasemonth_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
sns.color_palette('Spectral', 12)
```

```python
cp = sns.color_palette("Spectral", 12)
cp[0], cp[1], cp[2], cp[3], cp[4], cp[5], cp[6], cp[7], cp[8], cp[9], cp[10], cp[11] = cp[11], cp[9], cp[7], cp[5], cp[3], cp[1], cp[0], cp[2], cp[4], cp[6], cp[8], cp[10]
cp
```

```python
cat2 = sns.catplot(x = releasemonth_names, y = releasemonth_values, kind = 'bar',
                   hue = releasemonth_names, palette = cp, legend = True)
plt.show()
```

```python
rel1 = sns.relplot(data = spot, x = 'energy_%', y = 'bpm')
plt.show()
```

```python
modelist = spot.groupby('mode').count()['track_name']
modelist
```

```python
spot['month_cat'] = spot['released_month'].astype('category')
spot['month_cat'] = spot['month_cat'].cat.rename_categories(releasemonth_names)
```

```python
spot['month_cat'].head()
```

```python
rel2 = sns.relplot(data = spot, x = 'energy_%', y = 'bpm', hue = 'month_cat', palette = cp,
                   style = 'mode', markers = ['8', '>'], size = 'month_cat')
plt.show()
```

```python
rel3 = sns.relplot(data = spot, x = 'energy_%', y = 'bpm', col = 'mode', hue = 'mode', legend = False,
                   alpha = 0.5)
plt.show()
```

```python
rel4 = sns.relplot(data = spot, x = 'energy_%', y = 'bpm',
                   col = 'month_cat', col_wrap = 3,
                   hue = 'mode', style = 'mode')
plt.show()
```

```python
spot['release_season'] = spot['released_month'].replace({1: "Winter", 2: "Winter", 12: "Winter",
                                                    3: "Spring", 4: "Spring", 5: "Spring",
                                                    6: "Summer", 7: "Summer", 8: "Summer",
                                                    9: "Fall", 10: "Fall", 11: "Fall"}).astype('category')
```

```python
spot['release_season'].head()
```

```python
spot['release_season'].cat.reorder_categories(['Winter', 'Spring', 'Summer', 'Fall'], ordered = True)
```

```python
rel3 = sns.relplot(data = spot, x = 'energy_%', y = 'bpm', hue = 'release_season', palette = "Spectral",
                   style = 'mode', markers = ['8', '>'])
plt.show()
```
