```python
import pandas as pd
import plotly.express as px
import dash
import plotly.graph_objects as go
from dash import Dash, html, dcc, Output, Input, callback
```

```python
mb = pd.read_csv('mobile-use.csv')
mb.head()
```

```python
mb[(mb['Age'] > 20) & (mb['Age'] <= 40)]
```

```python
app1 = Dash()
app1.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.RangeSlider(20,
               90,
               step = 10,
               value = [20, 90],
               id = 'Age_slider')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value')
)
def update_fig(age_select):
    filtered_mb = mb[(mb['Age'] > age_select[0]) & (mb['Age'] <= age_select[1])]
    fig = px.scatter(filtered_mb, x = 'spAge', y = 'spComfort', color = 'Sex')
    return fig


if __name__ == '__main__':
    app1.run(debug = True)
```

```python
app2 = Dash()
app2.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.RangeSlider(20,
               90,
               step = 10,
               value = [20, 90],
               id = 'Age_slider')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value')
)
def update_fig(age_select):
    filtered_mb = mb[(mb['Age'] > age_select[0]) & (mb['Age'] <= age_select[1])]
    men_mb = filtered_mb[filtered_mb['Sex'] == 1]
    women_mb = filtered_mb[filtered_mb['Sex'] == 2]
    
    trace_men = go.Scatter(x = men_mb['spAge'], y = men_mb['spComfort'],
                           mode = 'markers', name = 'Men')
    trace_women = go.Scatter(x = women_mb['spAge'], y = women_mb['spComfort'],
                             mode = 'markers', name = 'Women')
    fig = go.Figure(data = [trace_men, trace_women])
    # fig = px.scatter(filtered_mb, x = 'spAge', y = 'spComfort', color = 'Sex')
    return fig


if __name__ == '__main__':
    app2.run(debug = True)
```
