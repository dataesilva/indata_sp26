### Reaction mojibake
[Mojibake](https://en.wikipedia.org/wiki/Mojibake) is the result of bed character encoding. It is shown as non-senseable characters that replace or mask the intended characters. To fix mojibake issues, often you just need to import or export text data while specifying the desired encoding scheme. Python (and most of the web) use UTF-8 encoding, so this is usually a safe option. However, programs like Excel do not default to UTF-8 encoding, so this often needs to be specified.

In the `reactions` data, the issue is not in our side of encoding/decoding. Instead, Facebook encoded these poorly in their APIs, leaving us to fix the issue.

A very nice solution was provided by [Martijn Pieters](https://stackoverflow.com/questions/50008296/facebook-json-badly-encoded/50011987#50011987) to address this exact issue. The solution uses regular expressions (re) to find portions of text that match certain patterns and then defines an efficient function to decode and then encode correctly. The full explanation can be read on the original Stack Overflow answer.

```python
import re
from functools import partial
```

```python
fix_mojibake_escapes = partial(
    re.compile(rb'\\u00([\da-f]{2})').sub,
     lambda m: bytes.fromhex(m[1].decode()),
 )

with open("chessbuds_messages.json", "rb") as m:
    repaired = fix_mojibake_escapes(m.read())
dat_fixed = json.loads(repaired)
```

```python
mess_fixed = pd.DataFrame(dat_fixed['messages'])
mess_fixed.head()
```
