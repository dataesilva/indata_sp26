Plotly Intro

What is Plotly?
-Plotly is a free, open source library. It supports several languages including Pythons, JavaScript, R, and MATLAB. 
     
What is Plotly used for?
-It is used for data visualization, web-based graphics, and interactive dashboards.

Basic plotly charts and interactions (non-technical)
-Plotly supports several types of charts - basic, financial, statistical and more
-Charts available: 2D/3D scatter, line, bar, pie, heatmap, candlestick, funnel  and geospatial maps
-Interactions: hover tooltips, zooming, panning, click-events, and legend toggling for data exploration

Difference between Plotly graph objects (non-technical/just the basics) and Plotly Express
-Plotly Express is user-friendly and good for quick and simple charts. It is an in-built package that is less customizable.
-Plotly Graph Objects is for visualizations that are customized or more complex. It is also in-built but is more customizable.

# For all figures, I have given each a unique variable name. This will help, especially when updating traces and figures.

```python
status_counts = df['order_status'].value_counts().reset_index()
status_counts.columns = ['order_status', 'frequency']

# Create bar chart
fig1 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    title='Frequency of Order Status',
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'}
   
)

fig1.show()
```

```python
fig2 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    color='order_status',
    color_discrete_map={
        'Delivered': 'green',
        'Cancelled': 'red',
        'Delayed': 'yellow'
    },
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'},
    title='Frequency of Order Status',

)

#With legend to the right
fig2.show()
```

```python
fig3 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    color='order_status',
    color_discrete_map={
        'Delivered': 'green',
        'Cancelled': 'red',
        'Delayed': 'yellow'
    },
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'},
    title='Frequency of Order Status',

)

# Rename legend title
fig3.update_layout(legend_title_text='Legend')

#With legend to the right
fig3.show()
```

```python
fig4 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    color='order_status',
    color_discrete_map={
        'Delivered': 'green',
        'Cancelled': 'red',
        'Delayed': 'yellow'
    },
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'},
    title='Frequency of Order Status',

)
#Remove the Legend
fig4.update_layout(showlegend=False)

fig4.show()
```

```python
fig5 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    color='order_status',
    color_discrete_map={
        'Delivered': 'green',
        'Cancelled': 'red',
        'Delayed': 'yellow'
    },
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'},
    title='Frequency of Order Status',
)

#Remove the Legend
fig5.update_layout(showlegend=False)

fig5.show()
```

# Made additions to lines 6 and 17 below.
## The update_traces was correct, but missing was the text argument in the bar function. This is set in the same way as the other trace elements using the name of the column in the DataFrame. This will automatically associate the frequency data with the height of the bar (y) and the text attribute.
#### [Use of the `text` argument for bar charts in plotly express](https://plotly.com/python/bar-charts/#:~:text=The%20text%20argument%20can%20be%20used%20to%20display%20arbitrary%20text%20on%20the%20bars%3A)
## In addition, added an example of how to use update_traces to adjust text further. The template sets the style, static elements concatinated with the variable of the figure (text), which you set when calling the bar function.
### [More on templates and the %{variable} syntax](https://plotly.com/python/text-and-annotations/#:~:text=a%20Plotly%20annotation-,Customize%20Displayed%20Text%20with%20a%20Text%20Template,-To%20show%20an)
### - David

```python
fig6 = pe.bar(
    status_counts,
    x='order_status',
    y='frequency',
    color='order_status',
    text = 'frequency',
    color_discrete_map={
        'Delivered': 'green',
        'Cancelled': 'red',
        'Delayed': 'yellow'
    },
    labels={'order_status': 'Order Status', 'frequency': 'Frequency'},
    title='Frequency of Order Status',
)

#Add Frequency count above each bar in black
fig6.update_traces(
    texttemplate = 'count = ' + '%{text}',
    textposition="outside",
    textfont = {'color': "black"}
)

#Remove the Legend
fig6.update_layout(showlegend=False)

fig6.show()
```
