```python
import pandas as pd
```

```python
com = pd.read_csv('commute.csv')
com.head()
```

```python
fig3, ax3 = plt.subplots()
ax3.plot(com['TripID'], com['CommuteTime'])

plt.show()
```

```python
fig4, axs4 = plt.subplots(2, 2)
axs4[1,1].plot(com['TripID'], com['CommuteTime'])

plt.show()
```

```python
fig5, axs5 = plt.subplots(5, 8)
axs5[2,2].plot(com['TripID'], com['CommuteTime'])
axs5[1,7].plot(com['TripID'], com['Date'])
axs5[4, 2].plot(com['TripID'], com['Destination'])

plt.show()
```

```python
fig6, axs6 = plt.subplots(2, 1)
axs6[1].plot(com['TripID'], com['CommuteTime'])
plt.show()
```

```python
fig7, axs7 = plt.subplots(1, 2)
axs7[1].plot(com['TripID'], com['CommuteTime'])
axs7[1].plot(com['TripID'], com['Date'])
axs7[1].set_xlabel('Trip Identity')
axs7[1].set_ylabel('Date and Time')

plt.show()
```

```python
fig8, ax8 = plt.subplots()
ax8.hist(com['CommuteTime'], bins = 3, color = 'cyan')

plt.show()
```

```python
fig9, ax9 = plt.subplots()
ax9.scatter(com['CommuteTime'], com['Vehicle'])

plt.show()
```
