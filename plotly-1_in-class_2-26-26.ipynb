```python
import plotly as pl
import plotly.express as px
import pandas as pd
```

```python
fdo = pd.read_csv('daily_food_delivery_orders.csv')
fdo.head()
```

```python
status_counts = fdo['restaurant_type'].value_counts().reset_index()
status_counts.columns = ['restaurant_type', 'frequency']
fig1 = px.bar(status_counts,
              x = 'restaurant_type',
              y = 'frequency',
              title = "frequency of Restaurant Type",
              labels = {'restaurant_type': "Restaurant Type", 'frequency': 'Count'},
              text = 'frequency',
              color = 'restaurant_type',
              color_discrete_sequence = ['tan', '#D3885D', 'red', 'yellow', 'green', 'orange']
              )
fig1.show()
```

```python
fdo_20 = fdo.head(20)
fdo_20
```

```python
fig2 = px.line(fdo_20,
               x = 'order_date',
               y = 'order_value',
               markers = True)

fig2.show()
```

```python
fdo_20.loc[:,'datetime'] = pd.to_datetime(fdo_20['order_date'])
fdo_20.sort_values('datetime', inplace = True)
```

```python
fig3 = px.line(fdo_20,
               x = 'datetime',
               y = 'order_value',
               markers = True,
               labels = {'datetime': 'Order Date',
                         'order_value': 'Order Value'})

fig3.update_traces(marker = {'color': 'green',
                             'size': 10,
                             'symbol': 'square'},
                    line = {'width': 3, 'color': 'red'})

fig3.show()
```

```python
fig4 = px.scatter(fdo,
                  x = 'delivery_distance_km',
                  y = 'delivery_time_minutes',
                  color = 'delivery_partner_rating',
                  symbol = 'order_status')

fig4.show()
```

```python
fdo_samp = fdo.sample(n = 50)
fdo_samp.head()
```

```python
fig5 = px.scatter(fdo_samp,
                  x = 'delivery_distance_km',
                  y = 'delivery_time_minutes',
                  color = 'delivery_partner_rating',
                  symbol = 'order_status')

fig5.update_traces(marker = {'size': 20, 'opacity': 0.5,
                             'line': {'width': 2, 'color': 'black'}})

fig5.update_layout(legend = {'yanchor': 'top', 'y': 0.98, 'xanchor': 'left', 'x': 1.23})

fig5.show()
```
