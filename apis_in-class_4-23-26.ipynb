```python
import requests
import json
import urllib
import time
import pandas as pd
```

```python
met_api = 'https://collectionapi.metmuseum.org/'
objects_endpoint = 'public/collection/v1/objects'
departments_endpoint = 'public/collection/v1/departments'
object_endpoint = 'public/collection/v1/objects'
search_endpoint = 'public/collection/v1/search'
```

```python
met_api+objects_endpoint
```

```python
objects_result = requests.get(met_api+objects_endpoint)
objects_result.status_code
```

```python
objects_result.json().keys()
```

```python
departments_result = requests.get(met_api+departments_endpoint)
departments_result.status_code
```

```python
helmets_query = '?q={}&departmentId={}'.format('helmet', 4)
met_api+search_endpoint+helmets_query
```

```python
helmets_response = requests.get(met_api+search_endpoint+helmets_query)
helmets_response.status_code
```

```python
helmets_response.json()
```

```python
met_api+object_endpoint
```

```python
helmet_data = []
for i in helmets_response.json()['objectIDs']:
    helmet_response = requests.get(
        met_api+object_endpoint+'/{}'.format(str(i)))    
    if helmet_response.status_code == 200:
        helmet_data.append(helmet_response.json())
    else:
        helmet_data.append('uhoh')
    time.sleep(1)
    # print(i)

helmet_data[0]
```
