```python
query='?departmentId={}&q={}'.format(4, 'helmet')
query
```

```python
helmet_response = requests.get(api_stem+query)
helmet_response.status_code
```

```python
helmet_response.json().keys()
```

```python
helmets = []
for i in helmet_response.json()['objectIDs'][0:10]:
    object_endpoint = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/'
    helmets_response = requests.get(object_endpoint+str(i))
    helmets.append(helmets_response.json())
    time.sleep(.5)
    print(helmets_response.status_code)
```

```python
helmets[0].keys()
```

```python
helmets_df = pd.DataFrame(helmets)
helmets_df.head()
```
